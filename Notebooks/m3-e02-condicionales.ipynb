## Tipo de dato bool

Es un tipo de dato boolean

1 == True
0 == False

```python
verdadero = True
print(verdadero)
print(type(verdadero))
```

```python
falso = False
print(falso)
print(type(falso))
```

## Estructura condicional if

```python
if verdadero:
    print("La prueba es verdadero")
else:
    print("La prueba es falso")
```

## Operadores relacionales

~~~ text
== igualdad
> mayor que
< menor que
>= mayor o igual que
<= menor o igual que
!= Diferente
~~~~


```python
print(10 == 10)
print( f"Operador de comparación igualdad: {10 == 10}" )

print("Operador de comparación igualdad: " + str(10 == 10) + " " , type(str(10 == 10)) )

```

```python
print( f"{ (10 == 10) = }")
print( f"{ (10 > 10) = }")
print( f"{ (10 < 10) = }")
print( f"{ (10 >= 10) = }")
print( f"{ (10 <= 10) = }")
print( f"{ (10 != 10) = }")
```

### Se necesita desarrollar una validación si una persona es mayor de edad

```python
edad = 18
if edad >= 18:
    print("La persona es mayor de edad")
else:
    print("La persona es menor de edad")
```

## Operadores lógicos

~~~ text
and   Devuelve verdarero si todas son verdad
or    Devuelve verdarero si al menos una es verdad
not   Negación
~~~

```python
print ( f"{(True and True) = }" )
print ( f"{(True and False) = } " )
print ( f"{(False and True) = } " )

print ( f"{(True or True) = }" )
print ( f"{(True or False) = } " )
print ( f"{(True or False) = } " )

print ( f"{(not False) = } " )
print ( f"{(not True) = }" )
```

### Desarrollar una condición para saber si una persona esta autorizada para conducir un vehículo. Asumir que la persona debe ser mayor de edad y licencia de conducir

### Para tener acceso a nuestro sistema, el usuario no puede estar bloqueado en la base de datos de ingresos permitidos.

```python
usuario_bloqueado = False

if not usuario_bloqueado:
    print("Bienvenido")
else:
    print("Acceso no permitido")
```

### Comparación de tipos de datos

variable

- Nombre
- Valor
- Tipo de dato

```python
# Los tipos de datos en Python son objetos

nombre = "German"

print( f"La variable nombre contiene {nombre} y es de tipo {type(nombre)}")
print (nombre.upper() )
```

```python
# No se puede comparar un class (objeto) con una cadena de texto
# <class 'str'> == "str"
if type(nombre) == "str":
    print("La variable es un string (Cadena de texto)")
else:
    print("No se puede comparar")
```

```python
if type(nombre) == str:
    print("La variable es un string (Cadena de texto)")
else:
    print("No se puede comparar")
```

```python
# Forma recomendada
# Traducción: si la variable nombre es una instacia de la clase str
if isinstance(nombre, str):
    print("La variable es un string (Cadena de texto)")
else:
    print("No se puede comparar")
```
