# Sentencias iterativas



## El problema

Se requiere imprimir la lista de notas de una en una.

```python
notas = (5, 6, 7, 4, 6,7)

print ( notas[0] )
print ( notas[1] )
print ( notas[2] )
print ( notas[3] )
```

```python
# Solución utilizar una estructura de control repetitiva (Ciclo/Bucle/sentencias iterativas)
for item_nota in notas:
  print( item_nota )
```

## Función range()

Genera una secuencia de números

```python
# Uso base
for i in range(5):
  print(i)
```

```python
# range() con número inicial
# Intervalo semi abierto
for i in range(2, 7):
  print(i)
```

```python
# range() con número inicial y step
for i in range(1, 20, 2):
  print(i)
```

## Palabras reservadas break y continue

Permiten cambiar el flujo de ejecución de nuestro ciclo

### break - detener la ejecución de un ciclo

Necesitamos buscar un número en una lista de números

```python
numeros = (1, 2, 3, 4, 5, 15, 6, 7)
buscar = 15

print ( f"Inicio buscando el número {buscar}")
for num in numeros:
  print( f"Analizando el número {num}: ", end="")
  if num == buscar:
    print("El número buscado fue encontrado")
    break
  else:
    print("El número no fue encontrado seguimos buscando")


print("---Fin del ciclo")
```

### continue - Regresa al inicio del ciclo

Imprimir los números pares según una lista generada por range()

```python
numeros_validar = range(20)
pares = []

for num in numeros_validar:
  if (num % 2) != 0:
    continue

  pares.append( num )

print( f"Los números pares son: {pares}")
```

## Tipo de dato dict (diccionario)

Los diccionarios son clave/valor y se encuentran entre llaves

### Ciclos for y los dict

```python
ventas = {
    "enero" : 15_000,
    "febrero" : 20_000,
    "marzo" : 22_000,
    "abril" : 20_000,
    "mayo" : 23_000
}

print (ventas)
```

```python
# Recorrido base

for mes in ventas: # Implicitamente estamos haciendo un ventas.keys()
  print( mes )
```

```python
# Acceder al valor - Forma 1

print("Ventas por mes")
for mes in ventas:
  print( f"Las ventas del mes {mes} son: {ventas[mes]}" ) # Acceso por clave



print("\nVentas por mes - Desempaquetado")
for mes, monto in ventas.items():
  print( f"Las ventas del mes {mes} son: {monto}" ) # Acceso desempaquetado
```

### Formatear salida en consola

El poder de print() y las f-string



```python
# Informe de ventas

print("Informe de ventas")
print ("=" * 17)
print ("\n")

print(f" {'Mes':<12} {'Ventas $':>15}")

for mes, monto in ventas.items():
  print( f"{mes:<12} {monto:>15,.0f}" )
```
