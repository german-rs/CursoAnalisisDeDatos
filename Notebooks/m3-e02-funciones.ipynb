### Mi primera función en Python

```python
def saludo():
  print("hola")


# Función con parámetros
def saludo_v2(nombre):
  print( f"Hola {nombre} ¿Cómo estás?")


# Función con parámetro y valor predeterminado
def saludo_v3(nombre = "Desconocido"):
  print( f"Hola {nombre} ¿Cómo estás?")


# Función con parámetro, valor predeterminado y declaración de tipo
def saludo_v4(nombre : str = "Desconocido") -> str:
  print( f"Hola {nombre} ¿Cómo estás?")
```

```python
#saludo()
saludo_v2("Juan")

saludo_v3()
saludo_v3("Belén")

saludo_v4()
saludo_v4("German")
saludo_v4(123)
```

```python
def imprimir_valor_tipo(valor: any) -> None:
  print(f"El valor es {valor} y su tipo es {type(valor)}")
```

```python
edad = 25
imprimir_valor_tipo(edad)

nombre = "Juan"
imprimir_valor_tipo(nombre)

imprimir_valor_tipo(25.2)

imprimir_valor_tipo([1, 2, 3])

imprimir_valor_tipo({"nombre" : "María José"})
```

## Función sumar 2 números con return

```python
def sumar(num1: int | float = 0, num2: int | float = 0) -> int | float:
  return num1 + num2
```

```python
r = sumar(10, 20.5)
print(r)
imprimir_valor_tipo(r)
```

## Función para calcular descuentos

```python
def calcular_descuento(precio: int | float, descuento: int | float) -> int | float:
  return precio - (precio * descuento / 100)




def calcular_descuento_v2(precio: int = 0, descuento_porc : float = 0) -> float:
  x = descuento_porc / 100
  return precio * (1 - x)

```

```python
res = calcular_descuento(1000, 15)
print(res)

res2 = calcular_descuento_v2(1000, 15)
print(res2)
```

## Sumar n números

```python
def sumar_v2(*numeros):
  imprimir_valor_tipo( numeros)
  return sum(numeros)
```

```python
print(sumar_v2(1, 2))
print(sumar_v2(1, 2, 3, 4, 5))
print(sumar_v2(1, 2, 3, 4, 5, 6, 7, 8, 9))
```
